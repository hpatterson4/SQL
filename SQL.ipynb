```python
sql = '''
SELECT r.name, AVG(f.calories) as avg_calories
FROM foods as f
INNER JOIN restaurants AS r ON r.id = f.restaurant_id
GROUP BY r.name
order by AVG(f.calories)
'''

df = pd.read_sql_query(sql, engine)
df
```

```python
# How to use aggregate as filter
# category name, average calories per category where avg caloroes > 700

sql = '''
SELECT c.name, AVG(f.calories)
FROM categories AS c
INNER JOIN categories_foods AS cf ON cf.category_id = c.id
INNER JOIN foods AS f ON cf.food_id = f.id
GROUP BY c.name
HAVING AVG(f.calories) > 700
ORDER BY AVG(f.calories) ASC

'''

df = pd.read_sql_query(sql, engine)
df
```
